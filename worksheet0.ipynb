```python
def unit_converter():
    """
    Converts units of Length, Weight, and Volume.

    User chooses the conversion type and inputs the value.
    The function returns the converted value.

    Returns:
        float: Converted result
    """

    try:
        print("Choose Conversion Type:")
        print("1. Length")
        print("2. Weight")
        print("3. Volume")

        choice = int(input("Enter choice (1/2/3): "))
        value = float(input("Enter value to convert: "))

        if choice == 1:
            print("1. Meters to Feet")
            print("2. Feet to Meters")
            option = int(input("Choose option: "))

            if option == 1:
                result = value * 3.28084
                print("Result:", result, "ft")
            elif option == 2:
                result = value / 3.28084
                print("Result:", result, "m")

        elif choice == 2:
            print("1. Kilograms to Pounds")
            print("2. Pounds to Kilograms")
            option = int(input("Choose option: "))

            if option == 1:
                result = value * 2.20462
                print("Result:", result, "lbs")
            elif option == 2:
                result = value / 2.20462
                print("Result:", result, "kg")

        elif choice == 3:
            print("1. Liters to Gallons")
            print("2. Gallons to Liters")
            option = int(input("Choose option: "))

            if option == 1:
                result = value * 0.264172
                print("Result:", result, "gal")
            elif option == 2:
                result = value / 0.264172
                print("Result:", result, "L")

        else:
            print("Invalid conversion type.")

    except ValueError:
        print("Invalid input. Please enter numeric values.")

unit_converter()
```

```python
def find_sum(numbers):
    """Returns the sum of numbers in the list."""
    return sum(numbers)

def find_average(numbers):
    """Returns the average of numbers."""
    return sum(numbers) / len(numbers)

def find_max(numbers):
    """Returns the maximum value."""
    return max(numbers)

def find_min(numbers):
    """Returns the minimum value."""
    return min(numbers)


try:
    print("Choose operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    choice = int(input("Enter choice: "))

    nums = list(map(float, input("Enter numbers separated by space: ").split()))

    if len(nums) == 0:
        print("List is empty!")

    elif choice == 1:
        print("Sum =", find_sum(nums))

    elif choice == 2:
        print("Average =", find_average(nums))

    elif choice == 3:
        print("Maximum =", find_max(nums))

    elif choice == 4:
        print("Minimum =", find_min(nums))

    else:
        print("Invalid choice")

except ValueError:
    print("Invalid input. Please enter numbers only.")
```

```python
def extract_every_other(lst):
    return lst[::2]

print(extract_every_other([1,2,3,4,5,6]))
```

```python
def get_sublist(lst, start, end):
    return lst[start:end+1]

print(get_sublist([1,2,3,4,5,6],2,4))
```

```python
def reverse_list(lst):
    return lst[::-1]

print(reverse_list([1,2,3,4,5]))
```

```python
def remove_first_last(lst):
    return lst[1:-1]

print(remove_first_last([1,2,3,4,5]))
```

```python
def get_first_n(lst,n):
    return lst[:n]

print(get_first_n([1,2,3,4,5],3))
```

```python
def get_last_n(lst,n):
    return lst[-n:]

print(get_last_n([1,2,3,4,5],2))
```

```python
def reverse_skip(lst):
    return lst[-2::-2]

print(reverse_skip([1,2,3,4,5,6]))
```

```python
def flatten(lst):
    result = []
    for i in lst:
        if isinstance(i,list):
            result.extend(i)
        else:
            result.append(i)
    return result

print(flatten([[1,2],[3,4],[5]]))
```

```python
def access_nested_element(lst,indices):
    element = lst
    for i in indices:
        element = element[i]
    return element

lst = [[1,2,3],[4,5,6],[7,8,9]]
print(access_nested_element(lst,[1,2]))
```

```python
def sum_nested(lst):
    total = 0
    for i in lst:
        if isinstance(i,list):
            total += sum_nested(i)
        else:
            total += i
    return total

print(sum_nested([[1,2],[3,[4,5]],6]))
```

```python
def remove_element(lst,elem):
    result = []
    for i in lst:
        if isinstance(i,list):
            result.append(remove_element(i,elem))
        elif i != elem:
            result.append(i)
    return result

print(remove_element([[1,2],[3,2],[4,5]],2))
```

```python
def find_max(lst):
    max_val = float('-inf')
    for i in lst:
        if isinstance(i,list):
            max_val = max(max_val,find_max(i))
        else:
            max_val = max(max_val,i)
    return max_val

print(find_max([[1,2],[3,[4,5]],6]))
```

```python
def count_occurrences(lst,elem):
    count = 0
    for i in lst:
        if isinstance(i,list):
            count += count_occurrences(i,elem)
        elif i == elem:
            count += 1
    return count

print(count_occurrences([[1,2],[2,3],[2,4]],2))
```

```python
def deep_flatten(lst):
    result = []
    for i in lst:
        if isinstance(i,list):
            result.extend(deep_flatten(i))
        else:
            result.append(i)
    return result

print(deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]))
```

```python
def average_nested(lst):
    flat = deep_flatten(lst)
    return sum(flat)/len(flat)

print(average_nested([[1,2],[3,4],[5,6]]))
```
